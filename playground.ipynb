Konwersja przedziałów wiekowych w string na dwie osobne kolumny. Dzięki temu można będzie operować na liczbach.

```python
df['age'].unique()
```

```python
def age_interval_to_numeric(age_str):
    if age_str.startswith('<'):
        return (0, int(age_str[1:]))
    if '>=' in age_str:
        return (int(age_str.replace('>=', '')), 100)
    if age_str == 'unknown':
        return (None, None) 
    min_age, max_age = age_str.split('-')
    return (int(min_age), int(max_age))

df['age_min'], df['age_max'] = zip(*df['age'].map(age_interval_to_numeric))
```

```python
df[['age', 'age_min', 'age_max']].sort_values(by=['age_min', 'age_max'])
```

```python
age_values  = df['age_min'].dropna().sort_values().unique().tolist()
age_values
```

Dodać kolumnę z tagami aby wartości 'Koty i Psy' były powiązane z 'Koty' oraz 'Psy'. 

```python
def split_animals(x):
    if 'Koty' in x and 'Psy' in x:
        return ['Koty', 'Psy']
    else:
        return [x]
    
df['fav_animal_tags'] = df['fav_animals'].apply(split_animals)
```

```python
tags = set(df['fav_animal_tags'].explode().unique())
tags
df['fav_animal_tags'].isna().sum()
```
